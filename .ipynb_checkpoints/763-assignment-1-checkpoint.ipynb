```python
import pandas as pd
import numpy as np
```

```python
##1. Read in the CSV file “FisherIris_ML.csv” into a matrix named as “Iris”. ##
```

```python
Iris = np.loadtxt("data/FisherIris_ML.csv", delimiter=",")

print(Iris)
```

```python
##2. Display the number of row and number of columns of the matrix “Iris”. ##
```

```python
print(Iris.shape)
```

```python
##3. Display all the row numbers that have the 5th column less than 0. ##
```

```python
print(Iris[(Iris[:,4]<0),:])
```

```python
##4. Remove the rows with the 5th column less than 0 from the “Iris” matrix. ##
```

```python
Iris = Iris[(Iris[:,4]>0),:]
print(Iris)
```

```python
##5. Display the number of row and number of columns of the “Iris” matrix again. ##
```

```python
print(Iris.shape)
```

```python
##6. Copy the first 4 columns in the new “Iris” matrix into a new matrix “X”. ##
```

```python
X = Iris[:,0:3].copy()
print(X)
```

```python
##7. Copy the 5th columns in the new “Iris” matrix into a new variable (or matrix) “Y”. ##
```

```python
Y = Iris[:,4].reshape(96,1).copy()
print(Y)
```

```python
##8. Display the maximum value and the minimum value of each column in “X”. ##
```

```python
print('X.max() = ',X.max(axis=0))
print('X.min() = ',X.min(axis=0))
```

```python
##9. Display the number of elements in the third column of the matrix “X” that is greater than 36. ##
```

```python
grt = X[(X[:,2]>36),:]
print(grt)
```

```python
print("Size of the array: ", grt.size)
print("Length of one array element in bytes: ", grt.itemsize)
print("Total bytes consumed by the elements of the array: ", grt.nbytes)
```
